```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('spx_7_9_20_248am.csv')
df
```

```python
df = df.rename(columns = {'Expiration Date':'Exp', 'Bid':'C.bid', 'Ask':'C.ask','Bid.1':'P.bid','Ask.1':'P.ask', 'IV':'C.IV', 'IV.1':'P.IV', 'Open Int':'C.OI', 'Open Int.1':'P.OI'})
```

```python
pull_date = pd.to_datetime('07/08/2020')
for date in pd.to_datetime(df['Exp'].unique()):
    if (date.weekday() == 4) and ((date - pull_date).days > 23):
        near_date = date
        break
for date in pd.to_datetime(df['Exp'].unique()):
    if (date.weekday() == 4) and ((date - near_date).days > 6):
        next_date = date
        break

print(near_date)
print(next_date)
```

```python
df['Exp'] = pd.to_datetime(df['Exp'])
near_month_df = df.loc[df['Exp'] == near_date].reset_index()
next_month_df = df.loc[df['Exp'] == next_date].reset_index()
```

```python
near_month_df
```

```python
near_month_df['C.mid'] = (near_month_df['C.bid'] + near_month_df['C.ask']) / 2
near_month_df['P.mid'] = (near_month_df['P.bid'] + near_month_df['P.ask']) / 2
next_month_df['C.mid'] = (next_month_df['C.bid'] + next_month_df['C.ask']) / 2
next_month_df['P.mid'] = (next_month_df['P.bid'] + next_month_df['P.ask']) / 2
```

```python
lowest_near = 100
for i in range(0, len(near_month_df)):
    if abs(near_month_df['C.mid'][i] - near_month_df['P.mid'][i]) < lowest_near:
        forward_near = near_month_df['Strike'][i]
        lowest_near = abs(near_month_df['C.mid'][i] - near_month_df['P.mid'][i])
lowest_next = 100
for i in range(0, len(next_month_df)):
    if abs(next_month_df['C.mid'][i] - next_month_df['P.mid'][i]) < lowest_next:
        forward_next = next_month_df['Strike'][i]
        lowest_next = abs(next_month_df['C.mid'][i] - next_month_df['P.mid'][i])
print(forward_near)
print(forward_next)
```

```python
near_month_df['P.bid'][0:5]
```

```python
#Determine strikes included
near_month_df['P.include'] = True
near_month_df['C.include'] = True
next_month_df['C.include'] = True
next_month_df['P.include'] = True
for i in range(len(near_month_df) - 1, 0, -1):
    if (near_month_df['P.bid'][i] == 0):
        if (near_month_df['P.bid'][i-1] == 0):
            near_month_df['P.include'][0:i] = False
        else:
            near_month_df['P.include'][i] = False
for i in range(len(next_month_df) - 1, 0, -1):
    if (next_month_df['P.bid'][i] == 0):
        if (next_month_df['P.bid'][i-1] == 0):
            next_month_df['P.include'][0:i] = False
        else:
            next_month_df['P.include'][i] = False
for i in range(0, len(near_month_df) - 1):
    if (near_month_df['C.bid'][i] == 0):
        if (near_month_df['C.bid'][i+1] == 0):
            near_month_df['C.include'][i:len(near_month_df)] = False
        else:
            near_month_df['C.include'][i] = False
for i in range(0, len(next_month_df) - 1):
    if (next_month_df['C.bid'][i] == 0):
        if (next_month_df['C.bid'][i+1] == 0):
            next_month_df['C.include'][i:len(near_month_df)] = False
        else:
            next_month_df['C.include'][i] = False
```

```python
#Calculating T1 and T2
#pull_time = 2:48 AM
#both near and next term in this example are weekly expirations at 4:00 PM 
M_current = 12 + 21*60
M_settlement = 900
M_other_1 = ((near_date - pull_date).days - 2) * 1440
M_other_2 = ((next_date - pull_date).days - 2) * 1440
T_1 = (M_current + M_settlement + M_other_1) / 525600
T_2 = (M_current + M_settlement + M_other_2) / 525600
```

```python
#Setting rates
#https://www.treasury.gov/resource-center/data-chart-center/interest-rates/Pages/TextView.aspx?data=yield
R_1 = 0.000011
R_2 = 0.000011

F_1 = float(forward_near + np.exp(R_1 * T_1) * (near_month_df.loc[near_month_df['Strike'] == forward_near]['C.mid'] - near_month_df.loc[near_month_df['Strike'] == forward_near]['P.mid']))
F_2 = float(forward_next + np.exp(R_2 * T_2) * (next_month_df.loc[next_month_df['Strike'] == forward_next]['C.mid'] - next_month_df.loc[next_month_df['Strike'] == forward_next]['P.mid']))
```

```python
for i in range(0, len(near_month_df)):
    if near_month_df['Strike'][i] > F_1:
        break
    K_1 = near_month_df['Strike'][i]
for i in range(0, len(next_month_df)):
    if next_month_df['Strike'][i] > F_1:
        break
    K_2 = next_month_df['Strike'][i]
```

```python
calls_near = near_month_df.loc[(near_month_df['C.include'] == True) & (near_month_df['Strike'] > K_1)][['Strike', 'C.mid']].rename(columns = {'C.mid':'mid'})
puts_near = near_month_df.loc[(near_month_df['P.include'] == True) & (near_month_df['Strike'] <= K_1)][['Strike', 'P.mid']].rename(columns = {'P.mid':'mid'})
near_mids = puts_near.append(calls_near).reset_index(drop = True)
near_mids.loc[near_mids['Strike'] == K_1, 'mid'] = float((near_month_df.loc[near_month_df['Strike'] == K_1]['C.mid'] + near_month_df.loc[near_month_df['Strike'] == K_1]['P.mid']) / 2)
calls_next = next_month_df.loc[(next_month_df['C.include'] == True) & (next_month_df['Strike'] > K_2)][['Strike', 'C.mid']].rename(columns = {'C.mid':'mid'})
puts_next = next_month_df.loc[(next_month_df['P.include'] == True) & (next_month_df['Strike'] <= K_2)][['Strike', 'P.mid']].rename(columns = {'P.mid':'mid'})
next_mids = puts_next.append(calls_next).reset_index(drop = True)
next_mids.loc[next_mids['Strike'] == K_2, 'mid'] = float((next_month_df.loc[next_month_df['Strike'] == K_2]['C.mid'] + next_month_df.loc[next_month_df['Strike'] == K_2]['P.mid']) / 2)
```

```python
near_mids['delta_K'] = 0
near_mids['delta_K'][0] = near_mids['Strike'][1] - near_mids['Strike'][0]
near_mids['delta_K'][len(near_mids)-1] = near_mids['Strike'][len(near_mids)-1] - near_mids['Strike'][len(near_mids)-2]
for i in range(1, len(near_mids)-1):
    near_mids['delta_K'][i] = (near_mids['Strike'][i + 1] - near_mids['Strike'][i - 1]) / 2
next_mids['delta_K'] = 0
next_mids['delta_K'][0] = next_mids['Strike'][1] - next_mids['Strike'][0]
next_mids['delta_K'][len(next_mids)-1] = next_mids['Strike'][len(next_mids)-1] - next_mids['Strike'][len(next_mids)-2]
for i in range(1, len(next_mids)-1):
    next_mids['delta_K'][i] = (next_mids['Strike'][i + 1] - next_mids['Strike'][i - 1]) / 2
```

```python
var_1 = ((2/T_1)*np.exp(R_1*T_1)*np.sum(near_mids['mid'] * near_mids['delta_K'] / np.square(near_mids['Strike'])) - ((1/T_1) * ((F_1/K_1) - 1)**2))
var_2 = ((2/T_2)*np.exp(R_2*T_2)*np.sum(next_mids['mid'] * next_mids['delta_K'] / np.square(next_mids['Strike'])) - ((1/T_2) * ((F_2/K_2) - 1)**2))
```

```python
N_T1 = M_current + M_settlement + M_other_1
N_T2 = M_current + M_settlement + M_other_2
N_30 = 43200
N_365 = 525600
VIX = 100 * ((T_1*var_1*(N_T2 - N_30)/(N_T2 - N_T1) + T_2*var_2*(N_30 - N_T1)/(N_T2 - N_T1))*(N_365/N_30))**0.5
```
